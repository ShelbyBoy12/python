```python
# finding the 13 adjacent digits with largest product
i = 0
m = 1
while i < 1000 :
    x = l[i:(13+i)]   # slicing the list into 13 digits
    p = 1
    for j in x :
        p = p*j  # then finding their product
    if p > m :
        m = p  # if product is larger than the previous one, changing the product
    i = i + 1
print(m)
```
